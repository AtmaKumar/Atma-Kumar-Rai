Polynomial Regression

```
import pandas as pd
```

```
data=pd.read_csv("poly1.csv")
```

```
data.head()
```

```

x=data["x"]
y=data["y"]
```

```
import matplotlib.pyplot as plt
plt.plot(x,y)
```

```
import numpy as np
x=np.array(x)
y=np.array(y)
```

```
w0=.1
w1=.2
w2=.3
```

```
yp=w0+w1*x+w2*x**2
```

```
j=(sum((y-yp)**2))/13
```

```
j
```

```
k=0
alpha=.0001
while(k<3000):
    k=k+1
    yp=w0+w1*x+w2*x**2
    j=(sum((y-yp)**2))/13
    print(j)
    dw0=(sum(y-yp))/13
    dw1=(sum((y-yp)*x))/13
    dw2=(sum((y-yp)*(x**2)))/13
    w0=w0+alpha*dw0
    w1=w1+alpha*dw1
    w2=w2+alpha*dw2
    
```

```
x1=x["Interest Rate"]
x2=x["Unemployment Rate"]
y=x["Stock Price"]
```

```

print("length",len(x1))
print("length",len(x2))

print("length",len(y))

x1=np.array(x1)
x2=np.array(x2)
y=np.array(y)

```

```
lr=.01
k=0
while(k<40000):
    yp=w0+w1*x1+w2*x2
    j=sum((yp-y)**2)/27
    print(j)
    dw0=sum(yp-y)/27
    dw1=sum((yp-y)*x1)/27
    dw2=sum((yp-y)*x2)/27
    w0=w0-lr*dw0
    w1=w1-lr*dw1
    w2=w2-lr*dw2
    k=k+1
```
